# Test Visualization of MIP

```python
import matplotlib.pyplot as plt
from max_intensity_proj import create_mip, create_mip_montage, create_inverted_mip, create_inverted_mip_montage
```

```python
NIFTI_PATH =  "/Users/leonardvincentramil/Desktop/Semester Project/ITKTubeTK/ITKTubeTK-MRA/Normal099-MRA.nii.gz"
MASK_PATH = "/Users/leonardvincentramil/Desktop/Semester Project/ITKTubeTK/Segmentation/SPOCKMIP/Normal099-MRA.nii.gz"
PROJ_DIM = "axial"  # o "sagittal" o "coronal"
```

## MIP Image (Decide the Projection Dimension)

```python
mip_image = create_mip(nifti_filepath=NIFTI_PATH, 
           mask_filepath=MASK_PATH,
           projection_dimension=PROJ_DIM)
```

```python
plt.figure(figsize=(6, 6)) 
plt.imshow(mip_image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
inverted_mip_image = create_inverted_mip(nifti_filepath=NIFTI_PATH, 
           mask_filepath=MASK_PATH,
           projection_dimension=PROJ_DIM)
```

```python
plt.figure(figsize=(6, 6)) 
plt.imshow(inverted_mip_image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
plt.imsave('mip_image.png', mip_image, cmap='gray')
```
